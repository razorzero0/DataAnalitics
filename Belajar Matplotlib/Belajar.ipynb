```python
data_dict = {'Product A': 100,
             'Product B': 150,
             'Product C': 200,
             'Product D': 120,
             'Product E': 180}

# Membuat list dari kunci dan nilai dalam dictionary
products = list(data_dict.keys())
sales = list(data_dict.values())

# Membuat visualisasi
plt.figure(figsize=(8, 5))
plt.bar(products, sales, color='skyblue')

# Menambahkan label dan judul
plt.xlabel('Produk')
plt.ylabel('Penjualan')
plt.title('Penjualan Produk')

# Menampilkan plot
plt.xticks(rotation=45)  # Untuk memutar label sumbu x agar lebih terbaca
plt.tight_layout()  # Menyusun plot agar lebih rapi
plt.show()
```
